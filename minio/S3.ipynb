# Minio one Node

```python
import boto3

client = boto3.client(
    's3',
    endpoint_url='http://localhost:9000',
    aws_access_key_id='minioadmin',
    aws_secret_access_key='miniosecretkey'
)

```

```python
bucket_name = 'game-pictures'
client.create_bucket(Bucket=bucket_name)
```

```python
file_path = 'C:/Users/yanbe/Pictures/2.jpg'
object_name = '20133'
with open(file_path, 'rb') as file:
    client.put_object(Bucket=bucket_name, Key=object_name, Body=file)

# Проверка загрузки картинки
response = client.list_objects_v2(Bucket=bucket_name)
for obj in response['Contents']:
    print(obj['Key'])
```

```python
bucket_name = 'game-pictures'
file_name = '20133'

obj = client.get_object(Bucket=bucket_name, Key=file_name)
img_data = obj['Body'].read()
```

```python
import PIL.Image as Image
import io

img = Image.open(io.BytesIO(img_data))
```

```python
img
```
