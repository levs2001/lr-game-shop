```python
import pandas as pd
import numpy as np
import random
import string

data = {
    'game_id': [],
    'title': [],
    'developer': [],
    'publisher': [],
    'release_date': [],
    'genre': [],
    'platform': [],
    'rating': [],
    'image_key': []
}

for i in range(100):  
    game_id = np.random.randint(1000) 
    developer = ''.join(random.choices(string.ascii_uppercase, k=5))  
    publisher = ''.join(random.choices(string.ascii_uppercase, k=5))  
    release_date = np.random.choice(pd.date_range('2000-01-01', '2023-01-01'), size=1)[0] 
    genre = np.random.choice(['Action', 'Adventure', 'RPG', 'Strategy', 'Shooter'], size=1)[0] 
    platform = np.random.choice(['PC', 'PS4', 'Xbox One', 'Nintendo Switch', 'Mobile'], size=1)[0] 
    rating = np.random.uniform(0, 10, size=1)[0] 
    title = ''.join(random.choices(string.ascii_uppercase, k=5)) 
    image_key = f"{game_id}_{title}"  

    data['game_id'].append(game_id)
    data['title'].append(title)
    data['developer'].append(developer)
    data['publisher'].append(publisher)
    data['release_date'].append(release_date)
    data['genre'].append(genre)
    data['platform'].append(platform)
    data['rating'].append(rating)
    data['image_key'].append(image_key)

df = pd.DataFrame(data)
```
